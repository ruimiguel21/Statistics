QUESTION 1:

We will state the null hypothesis:

*   $H\_0: β1$ = 0 (Charles river has no effect on the median houses)
*   $H\_1: β1$ is not equal to 0 (Charles river has effect on the median house)

```python
## X is the input variables (or independent variables)
X = boston_df['CHAS']
## y is the target/dependent variable
y = boston_df['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

Conclusion: For a 5% of significance level, we reject the null hyphotesis and therefore the Charles river has effect on the median houses, because the p-value e lower the 0.05.

QUESTION 2:

State the Hypothesis:

*   $H\_0:$ There is no difference in median home value between different proportions of owner-occupied units built before 1940
*   $H\_1:$ There is a difference in median home value between different proportions of owner-occupied units built before 1940

```python
boston_df.loc[(boston_df['AGE'] <= 35), 'age_group'] = '35 years and younger'
boston_df.loc[(boston_df['AGE'] > 35)&(boston_df['AGE'] < 70), 'age_group'] = 'between 35 and 70 years'
boston_df.loc[(boston_df['AGE'] >= 70), 'age_group'] = '70 years and older'
```

```python
from statsmodels.formula.api import ols
lm = ols('MEDV ~ age_group', data = boston_df).fit()
table= sm.stats.anova_lm(lm)
print(table)
```

Conclusion: We will reject the null hypothesis since the p-value is less than 0.05, so There is no difference in median home value (MEDV) between different proportions of owner-occupied units built before 1940
